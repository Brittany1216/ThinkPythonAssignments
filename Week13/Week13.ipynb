```python
# NAME: journal.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# 1. Define write_entry
def write_entry(filename, entry):
#    - Accept a filename and a string as parameters
#    - Close when finished
    with open('journal.txt', 'a') as file:
        file.write(entry + '\n')
    print('Entry written to journal.')

# 2. Define read_journal
#    - Read entire contents and print so users can see their past entries

# PROGRAM LOGIC
# 1. Ask user if they want to write a new entry or read their journal
#    - W for write a new entry and R for read journal
#    - If W ask for their thoughts and call write_entry
#    - If R call read_journal
# 2. Make a while loop so the user can keep using it until they type 'Quit'
if __name__ == "__main__":
    while True:
        print('--- Daily Journal ---')
        print('1. Write Entry')
        print('2. Exit')
        choice = input('Select an option (1-2): ')

        if choice == '1':
            user_entry = input('Enter your journal entry: ')
            write_entry('journal.txt', user_entry)
        elif choice == '2':
            break
        else:
            print("Invalid choice, try again.")
```
