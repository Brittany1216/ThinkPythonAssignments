```python
# NAME: rectangle.py
# IMPORT STATEMENTS


# FUNCTION DEFINITIONS
# 1. Define calaculate_rectangle
def calculate_rectangle(length, width):
    area = length * width
    perimeter = 2 * (length + width)
#    - Return single tuple in parameter of area and perimeter
    return area, perimeter


# PROGRAM LOGIC
# 1. Input length and width
while True:
    try:
        length = float(input("Enter length: "))
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value for length.")

while True:
    try:
        width = float(input("Enter width: "))
        break
    except ValueError:
        print("Invalid input. Please enter a numerical value for width.")
#    - Covert to float
area, perimeter = calculate_rectangle(length, width)
# 3. Two separate variables called my_area and my_perimeter
# 4. Print results
print(f"The Area is: {area}")
print(f"The Perimeter is: {perimeter}")
```
