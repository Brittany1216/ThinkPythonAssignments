```python
# NAME: roster.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# 1. Define add_players
def add_players(roster):
#    - Ask for players name
  while True:
    name = input("Enter player name (or 'quit' to stop): ")
    if name == 'quit':
      break
    roster.append(name)
  return roster
# 2. Define is_on_roster
def is_on_roster(roster, name):
#    - Return True if player is listed or false if player is not listed.
  if name in roster:
    return True
  else:
    return False

# PROGRAM LOGIC
# 1. Create empty list called Roster
team_roster = []
# 2. Call add_players to fill the list
team_roster = add_players(team_roster)
# 3. Ask user for specific name to search
player_name = input('Who would you like to search for? ')
# 3. Call is_on_roster
if is_on_roster(team_roster, player_name):
#    - Print 'Player Found!' or 'Player Not Found!'
  print('Result: Player Found!')
else:
  print('Result: Player Not Found!')
```
