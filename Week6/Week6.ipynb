```python
# IMPORT STATEMENTS
import math
# 1. Import math

# FUNCTION DEFINITIONS
def calculate_area(radius):
    return math.pi * radius**2
# 1. Define area(radius): returns pi * r squared
def calculate_circumference(radius):
    return 2 * math.pi * radius
# 2. Define circumference(radius): returns 2 * pi * r
def calculate_volume(radius, height):
    volume = calculate_area(radius) * height
    return volume
# 3. Define volume(radius, height):
#    - Call area(radius) and multiply by height.
#    - Return the result.

# PROGRAM LOGIC
r = float(input('Enter the radius: (Only numbers)'))
# 1. Input for radius (float)

area = calculate_area(r)
circumference = calculate_circumference(r)
print('Area:', area)
print('Circumference:', circumference)
# 2. Call and print area and circumference.

h = float(input('Enter the height: (Only numbers)'))
# 3. Input for height (float)

volume = calculate_volume(r, h)
print('Volume:', volume)
# 4. Call and print volume.
```
