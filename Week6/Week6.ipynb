```python
# IMPORT STATEMENTS
import math

# FUNCTION DEFINITIONS
def area(radius):
    return math.pi * radius**2

def circumference(radius):
    return 2 * math.pi * radius
# 3. Define volume(radius, height):
#    - Call area(radius) and multiply by height.
#    - Return the result.

# PROGRAM LOGIC
r = float(input('Enter the radius: (Only numbers)'))

area = area(r)
circumference = circumference(r)

print(f'Results for Radius: {r} ')
print(f'Area: {area:.4f}')
print(f'Circumference: {circumference:.4f}')

# 3. Input for height (float)
# 4. Call and print volume.
```
