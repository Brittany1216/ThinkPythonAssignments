```python
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# 1. Define is_palindrome
def is_palindrome(word):
  processed_word = str(word).lower()
  return processed_word == processed_word[::-1]


# PROGRAM LOGIC
# 1. Input word
user_input = input('Enter a word to check if it is a palindrome.')
# 2. Call users input
result = is_palindrome(user_input)
# 3. Print result
#    - Return True or False
if result:
  print(f"True '{user_input}' is a palindrome.")
else:
  print(f"False '{user_input}' is not a palindrome.")
```
