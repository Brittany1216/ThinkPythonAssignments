```python
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
# 1. Define verify_pin():
def verify_pin():
#    - Create a loop that asks for a PIN until it equals "1234"
  while True:
    pin = input('Enter pin:')
    if pin == '1234':
#    - Return True
      return True
      break
# 2. Define withdraw(balance, amount):
def withdraw(balance, amount):
#    - Check if enough money exists
    if amount > balance :
      print('Insufficient Funds.')
      return balance
    else:
      new_balance = balance - amount
#    - Return the new (or old) balance
      return new_balance

# PROGRAM LOGIC
# 1. Set current_balance to 500
balance = 500
# 2. Call verify_pin()
verify_pin()
# 3. Get withdrawal amount from user input
a = float(input('Enter withdraw amount:'))
# 4. Update current_balance by calling withdraw()
balance = withdraw(balance, a)
# 5. Print the final balance
print('Balance:', balance)
```
