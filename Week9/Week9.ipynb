```python
# NAME: rentalcar.py
# IMPORT STATEMENTS

# FUNCTION DEFINITIONS
# 1. Define car_rate(classification)
def car_rate(classification):
#    - Budget is 40 a day, Standard is 60 a day, and Luxury is 80 a day
  rates = {
        'B': 40.00,
        'S': 60.00,
        'L': 80.00
    }
  return rates.get(classification, 0.0) # Return the daily rate

# Helper function to get car classification name for display
def get_car_name(classification):
  car_class_names = {'B': 'Budget', 'S': 'Standard', 'L': 'Luxury'}
  return car_class_names.get(classification, 'Unknown')

# 2. Define cost_miles(miles_driven)
def cost_miles(miles_driven):
#    - Under 100 miles is free and anything over is 35 cents a mile
  if miles_driven < 100:
    return 0
  else:
    return (miles_driven - 100) * 0.35
# 3. Define distance traveled(start_odometer, end_odometer)
def distance_traveled(start_odometer, end_odometer):
#    - calculate the distanced traveled by subtracting starting odometer from end odometer)
  distance = end_odometer - start_odometer
  return distance
# 4. Define total_bill(classification, miles_driven, days)
def total_bill(classification, miles_driven, days):
#    - Call car_rate(classification) and multiply by cost_miles(miles_driven)
  daily_rental_cost = car_rate(classification) * days # Call car_rate to get the actual rate
  mileage_charge = cost_miles(miles_driven) # Pass the correct miles_driven argument
  total = daily_rental_cost + mileage_charge
  return total
# PROGRAM LOGIC
# 1. Input for classification
code = input("Enter classification (B, S, L): ").upper()
days = int(input("Enter number of rental days: "))
# 2. Input for miles_driven
start_odometer = int(input('Enter starting odometer reading: '))
end_odometer = int(input('Enter ending odometer reading: '))
miles = float(input("Enter total miles driven: "))
# 3. Call and print total_bill
calculated_total_bill = total_bill(code, miles, days) # Renamed variable to avoid conflict with function name
print("\n" + "="*30)
print(f"Classification: {get_car_name(code)}") # Use helper for name
print(f"Days: {days}")
print(f"Starting Odometer: {start_odometer}")
print(f"Ending Odometer: {end_odometer}")
print(f"Miles Driven: {miles}")
print(f"Total Charges: ${calculated_total_bill:.2f}") # Added formatting for currency
```
